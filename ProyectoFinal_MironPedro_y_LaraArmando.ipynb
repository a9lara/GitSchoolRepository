## [Modelos poblacionales](http://en.wikipedia.org/wiki/Epidemic_model#The_SIR_model) `SIS` y `SIR` 

Los modelos `SIS` y `SIR` son utilizados para modelar el comportamiento de epidemias, estos modelos son sencillos


En 1927  W.O Kermack y A.G. McKendrick crearon un modelo en el que consideran una población fija (para ver el trato original del problema epidemiologico se puede consultar [aquí](http://www.math.utah.edu/~bkohler/Journalclub/kermack1927.pdf)), distingiendo tres grupos en la población :

$S(t)$ $\rightarrow$ El número de individuos suceptibles a ser infectados

$I(t)$ $\rightarrow$ El número de individuos infectados 


$R(t)$ $\rightarrow$ El número de individuos que han sido infectados pero ya no estan enfermos (ya sea por inmunización o muerte) y no pueden contagiar a otros, ni contagiarse de nuevo.


La dinámica que intenta describir este modelo es:

$$ S \rightarrow I \rightarrow R $$

suponiendo una población fija, i.e. $N=S(t)+I(t)+R(t)$

Kermack y McKendrick derivaron las siguientes ecuaciones:

$$\frac{dS}{dt} = -\frac{\beta S I}{N} $$

$$\frac{dI}{dt} = \frac{\beta S I}{N} -\gamma I $$

$$\frac{dR}{dt} =  \gamma I$$

Varias supocisiones fueron hechas al formular las ecuaciones anteriores. Primero, que todo individuo en la población tiene la misma probabilidad de contagiarse con una ritmo $\beta$.

Por lo tanto un individuo infectado puede transmitir la enfermedad a otros $\beta N$ por unidad de tiempo y la fracción de interacciones entre un infectado y un suceptible es de $\frac{S}{N}$.

Dejando así el número de infecciones por unidad de tiempo  por infectado como $\beta N \frac{S}{N}$, resultando en un ritmo de infecciones nuevas de $\beta N (\frac{S}{N})I=\beta SI$.

Para las últimas dos ecuaciones se considera que la población que deja de ser suceptible pasa a ser infectada, de esta población de infectados cierta fracción  $\gamma$ por unidad de tiempo pasa de la clase de infectados a la clase de los removidos ($R$) .

Las razones que describe cada constante puede ser puesta como:

$\beta$ $\rightarrow$ Tasa de contagios 

$\frac{1}{\gamma}$ $\rightarrow$  Tiempo promedio de infección (para un solo individuo).

$\mu$ $\rightarrow$ Tasa promedio de defunciones y nacimientos.


- El modelo SIR con nacimientos y muertes

Un ejemplo es el caso del sarampión, en donde nuevos individuos llegan a la población y nacimientos y muertes deben ser tomados en cuenta. Las siguientes ecuaciones representan este modelo, suponiendo que la razón ($\mu$) de muertes y nacimientos es la misma 


$${dS \over dt} = - \frac{\beta S I}{N} + \mu (N-S)$$

$${dI \over dt} = \frac{\beta S I}{N} - \gamma I - \mu I$$

$${dR \over dt} = \gamma I - \mu R$$



- El modelo SIS con nacimientos y muertes

En el modelo SIS el individuo recobrado no desarrolla inmunidad a la enfermedad, por lo que lo vuelve de nuevo suscpetible y por lo tanto no hay recuperados, i.e. $R(t)=0$ para todo t. Y se describe por:

$${dS\over dt} = - \frac{\beta S I}{N} + \mu (N - S) + \gamma I $$

$${dI\over dt}= \frac{\beta S I}{N} - \gamma I - \mu I$$

```julia
k_1(x,t,f,h)=f(x,t)
k_2(x,t,f,h)=f(x+h*0.5*k_1(x,t,f,h),t+0.5*h)
k_3(x,t,f,h)=f(x+h*0.5*k_2(x,t,f,h),t+0.5*h)
k_4(x,t,f,h)=f(x+h*k_3(x,t,f,h),t+0.5*h)

function integrar(f,x0,t0,tf,h)
    
    rk4=Vector{Float64}[[x0]]
    t=Float64[t0]
    
            for i in t0:h:tf
        
                rk4_nueva=x0+h*(1/3*k_1(x0,t0,f,h)+1/6*k_2(x0,t0,f,h)+1/6*k_3(x0,t0,f,h)+1/3*k_4(x0,t0,f,h))
        
                t0=t0+h
                x0=rk4_nueva
        
        
                push!(t,t0)
                push!(rk4,[x0])
            
            end
        
        
         t,hcat(rk4...)'
            
        
    
end
```

## Modelación SIR sin dinamica de población

Para este modelo nos limitaremos a hacer un analísis de los parámetros $\gamma$ y $\beta$ para determinar bajo que condiciones la enfermedad no se convierte en una epidemia, considerando que para el tiempo inicial la poblacion total es de suceptibles (este análisis es completamente igual al hecho por Cammey Cole Manning en esta [presentación](http://www.math.utah.edu/~bkohler/Journalclub/kermack1927.pdf))i.e. $$S_0=N$$, y bajo que condiciones se establece una infección en dicha población que después de un tiempo pasa, en ambos casos para un tiempo finito debe pasar lo siguiente:

$$I = 0 $$

La poblacion de infectados es cero.

$$\frac{dR}{dt} = \gamma I $$


$$S+R=N$$

Para que la enfermedad se propague debe pasar que:

$$\frac{dI}{dt}>0$$

Como suponemos $S_0=N$, entonces la condición para el crecimiento se traduce en que:

$$ (\frac{\beta S_0}{N} - \gamma)I > 0 $$


$$ (\beta - \gamma)I > 0 $$

dividiento por $\gamma$,

$$ (\frac{\beta }{\gamma} - 1)I > 0 $$

Si denotamos a $\frac{\beta }{\gamma}$ como $R_0$, podemos interpretar esta constante como la que determina si la infección se propaga o no ya que si $R_0>1$  $\frac{dI}{dt}>0$, y la infección se propaga, pero si $R_0 < 1$ $\frac{dI}{dt}<0$ por lo que la infección muere y no hay epidemia, veamos si este comportamiento se ve reflejado en los resultados númericos.



### Implementación

Para resolver el método

$x[1] = S$

$x[2] = I$

$x[3] = R$

$x[4] = N$

$x[5] = \beta$

$x[6] = \gamma$

$x[7] = \mu$

```julia
function SIR(x,t)
    S=x[1]
    I=x[2]
    R=x[3]
    N=x[4]
    β=x[5]
    γ=x[6]
    μ=x[7]
    
    [-β*S*I/N , β*S*I/N - γ*I, γ*I,0,0,0,0]
end
```

Para el análisis fijamos $\gamma = 0.5$ y hacemos variar $\beta$, de igual manera hacemos que se pueda variar el número inicial de infectados ($\alpha$), mientras que suponemos que la población inicial de recuperados es cero, esto es razonable, pensando que estamos en un tiempo inicial tal que nadie se ha recuperado de la infección.

```julia
f = figure(figsize=(8,7))


@manipulate for β in 0:0.1:1.5, α in 0.01:0.1:0.91
        
        withfig(f) do

        x0=[1-α,α,0,1,β,0.5,0];

        tps1,resultados_sir=integrar(SIR,x0,0,50,0.1);

        suceptibles_sir=resultados_sir[:;1]
        infectados_sir=resultados_sir[:;2]
        inmunes_sir=resultados_sir[:;3]

        plot(tps1,suceptibles_sir,label="Susceptibles")
        plot(tps1,infectados_sir,label="Infectados")
        plot(tps1,inmunes_sir,label="Recuperados(Inmunes)")



        xlabel("Tiempo")
        ylabel("Población")
        title("SIR")
        legend()
        end
end
```

Al jugar con los valores de $\beta$ podemos observar que coincide completamente con la descripción dada al inicio de esta sección, ya que para $\beta \leq 0.5$ no se observa el crecimiento de la población de infectados, mientras que para $\beta > 0.5$ i.e 
$R_0 >1$ tenemos que la población presenta un máximo de individuos infectados para después desaparecer en un tiempo finito.

### Aplicando los modelos Epidemiológicos que incluyen dinámica de poblaciones

Vamos a suponer dos grupos de personas, el Grupo A y el Grupo B. 

Grupo A - El primer grupo serán personas que no tienen acceso a servicios médicos pues no habrá recuperados (las personas se vuelven a enfermar).

Grupo B - El segundo grupo serán individuos  que tienen acceso a servicios médicos lo que implica ciertas medidas para que los que ser recuperen no vuelvan a enfermarse por lo que en este grupo aplicaremos el modelo SIR (las personas se recuperan y viven). 

Para poder resolver estas ecuaciones diferenciales ordinarias utilizaremos el método de integración Runge-Kutta de cuarto grado.


### Modelación SIS con dinámica de población

Para este caso sólo consideramos las siguientes condiciones iniciales


$S(0)=815 890$ 

$I(0)=9 407$

$R(0)=4 703$


Con las constantes:


$N=830 000$

$\beta=0.7$ 

$\gamma=0.1$ 

$\mu=0.40$ 


```julia
function SIS_P(x,t)

    S=x[1]
    I=x[2]
    R=x[3]
    N=x[4]
    β=x[5]
    γ=x[6]
    μ=x[7]

    [-β*S*I/N + μ*(N-S) + γ*I, β*S*I/N-γ*I-μ*I,0,0,0,0,0]
end


function SIR_P(x,t)
    S=x[1]
    I=x[2]
    R=x[3]
    N=x[4]
    β=x[5]
    γ=x[6]
    μ=x[7]
    
    [-β*S*I/N + μ*(N-S), β*S*I/N - γ*I-μ*I, γ*I-μ*R,0,0,0,0]
end
```

### Resultados 

```julia
x1=[815890,9407,4703,830000,0.7,0.1,0.40];

tps2,resultados_sisp=integrar(SIS_P,x1,0,100,0.1);

suceptibles_sisp=resultados_sisp[:;1]
infectados_sisp=resultados_sisp[:;2]

plot(tps2,suceptibles_sisp,label="Susceptibles")
plot(tps2,infectados_sisp,label="Infectados")



xlabel("Tiempo")
ylabel("S(t),I(t)")
title("SIS ")
legend()
```

Aunque el contraste con el ejemplo anterior no puede hacerse directo (ya que se tratan de diferentes modelos), lo que si se puede observar es que la pobación de infectados parece estabilizarse para un número diferente de cero, este resultado puede interpretarse como una consecuencia del ingreso de un número constante de nuevos suceptibles y a su vez de un número constante de infectados, las curvas describen de que forma las diferentes poblaciones cambian antes de llegar a un estados de "equilibrio", es decir cuando las poblaciones de ambos tipos no presentan ningún cambio.

### Modelación SIR con dinámica de población

```julia
x2=[815890,9407,4703,830000,0.7,0.1,0.40];

tps3,resultados_sirp=integrar(SIR_P,x2,0,100,0.1);

suceptibles_sirp=resultados_sirp[:;1]
infectados_sirp=resultados_sirp[:;2]
inmunes_sirp=resultados_sirp[:;3]

plot(tps3,suceptibles_sirp,label="Susceptibles")
plot(tps3,infectados_sirp,label="Infectados")
plot(tps3,inmunes_sirp,label="Recuperados(Inmunes)")



xlabel("Tiempo")
ylabel("S(t),I(t),R(t)")
title("SIR")
legend()
```

Estos resultados presentan las mismas características que el anterior, solo que tenemos una tercera población (inmunes) que tiene el mismo comportamiento que las demás, un cambio continuo desde las condiciones iniciales a las de equilibrio.

### Comparación de SIR y SIS con población

```julia
plot(tps2,suceptibles_sisp,label="Susceptibles - SIS")

plot(tps3,suceptibles_sirp,label="Susceptibles - SIR")

xlabel("Tiempo")
ylabel("S(t)")
title("SIS vs SIR // Susceptibles")
legend()
```

En esta gráfica se observa una muy pequeña diferencia entre los dos moldelos cuando consideramos solo la población de suceptibles, esta diferencia es explicada por el término extra en la ecuación del modelo `SIS` (+$\gamma$ I), haciendo que la velocidad de descenso tenga menos inclinación que la población modelada por `SIR`.

```julia

plot(tps2,infectados_sisp,label="Infectados - SIS")
plot(tps3,infectados_sirp,label="Infectados - SIR")
xlabel("Tiempo")
ylabel("I(t)")
title("SIS vs SIR // Infectados")
legend()
```

En cambio para esta gráfica se observa el efecto que tiene la inclusión de la población extra de Inmunes en el modelo `SIR` en comparación con `SIS`, lo que hace que parte de la población de los infectados se desplacen a la de inmunes, lo cual se ve claramente en la diferencia de poblaciones.

### Ondas en el sistema de Reacción-Difusión de Rabia (Sin dinámica de Población)

El hecho de no tomar en cuenta la dinámica de Población en el grupo de perros suceptibles (S) quiere decir que el cambió en su población solo depende de la interacción con los infectados, mientras que la poblacion de perros infectados depende de la interaccion con los suceptibles, que tan rapido se mueran y del término de difusión, es decir:


$$\frac{dS}{dt}=-rIS$$

$$\frac{dI}{dt}=rIS -aI-D\frac{d^2 I}{dx^2}$$


Adimensionalizando esta ecuación al introducir $m=\frac{a}{rS_0}$ y para que podamos resolver este sistema con nuestro método de integración consideramos que la solución es una onda viajera, es decir

$$I=I(x-ct)=I(z)$$

$$S=S(x-ct)=S(z)$$

Y con la regla de la cadena, llegamos a las siguientes ecuaciones en función de la variable z:

$$ cS'= IS$$

$$ I'' - cI + (S-m)I= 0$$

Introducimos las siguientes condiciones de frontera al suponer que muy lejos la población de zorros infectados desaparece, y solo queda la poblacion de zorros suceptibles.


$$S(+ \infty)=1$$


$$I(- \infty)=0$$

Al también suponer que después del paso de la onda no hay fuentes de infeccion ,ya que la mayoría de los casos de rabia son fatales, y que la población de zorros se termina por estabilizar tambien tenmos las condiciones:

$$I(-\infty)=0$$

y 


$$S'(-\infty)=0$$


Estas consideraciones son analizadas en las [notas](http://www-m6.ma.tum.de/~kuttler/script_reaktdiff.pdf) de Christina Kuttler n las paginas 65 y 66, y dan como resultado que c debe complir con que $c\geq2\sqrt{1-m}$


introduciendo la variable $I'=v$, tenemos el sistema:

$$I'=v$$

$$S'=\frac{IS}{c}$$

$$v'=-cv-(S-m)I$$

Ecuaciones que traducimos en una función vectorial para que sean resueltas por el métododo de Runge-Kutta.

Ya que en un eje tenemos $z=x-ct$, podemos interpretar estas graficas como el frente de onda que barre el espacio para diferentes tiempos, ya que dejando una x fija y haciendo que el tiempo corra podemos ver como la figura se va desplazando por el espacio de izquierda a derecha con la misma forma, en este caso vemos que para un punto a una distancia, digamos a 10 unidades de distancia  de donde se inició la epidemia de rabia tenemos una población inicial que es cerca a $3.3$ que al paso de la epidemia pierde parte de su población hasta llegar a $1$, mientras que en el caso de la población rabiosa para un tiempo inicial en $x=10$ es cero, pero con el paso del tiempo esta incrementa hasta llegar a un máximo para despues decrecer y dejar a cero infectados despues de su paso.

#### Problemas con esto

Aunque la condición inicial de nuestro sistema fue una población de casi $1$ de acuerdo con la interpretación anterior se tiene una población inicial de alrededor de $3.3$ por esto podemos decir que la implementación del modelo (por nuestra parte) no es buena, además, esta gráfica no nos dice nada sobre el comportamiento de las poblaciones después del paso de la rabia, ya que se corta cuando la población de no infectados llega a $0.9983$ y la de infectados llega a $0.0013$( esto lo sabemos por como se escribió el código), lo que nos lleva a reforzar el hecho que es la mala implementación del modelo.
