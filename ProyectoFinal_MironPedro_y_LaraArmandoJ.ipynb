###  Modelos Epidemiológicos

Para poder describir la propagación de enfermedades que atacan a una población se pueden usar modelos basados en un sistema de ecuaciones diferenciales.

Cada ecuación nos proporciona información sobre las diferentes etapas que tiene la enfermedad y su distribución en la población.

La diferencia entre los modelos epidémicos y endémicos es que los primeros se basan en un brotes repentinos de enfermedades y los endémicos para enfermedades permanentes.

Vamos a mencionar dos modelos, el SIS y el SIR en su forma determinista y con dinámica pobaclional (vamos a considera nacimientos a lo largo del tiempo ''t'', de modo que hay una constante renovación de individuos susceptibles a la enfermedad. Del mismo modo, algunos individuos mueren o bien por la enfermedad o de manera natural al cabo de un tiempo de haber sido recobrados). En ambos se asume que la interacción entre los individuos es aleatoria. Además se considera en ambos casos el número total de individuos como $N=S(t)+I(t)+R(t)$. 

Los parámetros a usar serán los siguientes:

- Parámetros

$S$ $\rightarrow$ Perros Suceptibles a ser infectados 

$I$ $\rightarrow$ Perros Infectados 

$R$ $\rightarrow$  Perros Inmunes

$N$ $\rightarrow$  Población Total de Perros

$\beta$ $\rightarrow$ Tasa de contagios 

$\frac{1}{\gamma}$ $\rightarrow$  Tiempo promedio de infección (para un solo individuo).

$\mu$ $\rightarrow$ Tasa promedio de defunciones (probabilidad de que un individuo infectado muera debido a la enfermedad).






- SIS

En el modelo SIS el individuo recobrado no desarrolla inmunidad a la enfermedad, por lo que lo vuelve de nuevo suscpetible y por lo tanto no hay recuperados, i.e. $R(t)=0$ para todo t. Y se describe por:

$${dS\over dt} = - \beta S I + \mu (N - S) + \gamma I $$

$${dI\over dt}= \beta S I - \gamma I - \mu I$$

- SIR

En este caso las infecciones llevan a una inmunidad vitálica, es decir, existe la recuperación del individuo. Y se describe por:

$${dS \over dt} = - \beta S I + \mu (N-S)$$

$${dI \over dt} = \beta S I - \gamma I - \mu I$$

$${dR \over dt} = \gamma I - \mu R$$




### Aplicando los modelos Epidemiológicos

Vamos a suponer dos grupos de perros, el Grupo A y el Grupo B. 

Grupo A - El primero grupo serán perros callejeros los cuáles no tienen dueños que los lleven al veterinario. Por ello se usará el modelo SIS, pues no habrá recuperados.

Grupo B - El segundo grupo serán perros con dueños en los cuáles, gracias a la pronta respuesta de sus dueños, el veterinario les puede curar. En este grupo aplicaremos el modelo SIR. 

Para poder usar estas ecuaciones diferenciales ordinarias con un método de integración vamos a implementar RungeKutta4.

### Implementación

$x[1] = S$

$x[2] = I$

$x[3] = R$

$x[4] = N$

$x[5] = \beta$

$x[6] = \gamma$

$x[7] = \mu$

```julia
using PyPlot
```

```julia
k_1(x,t,f,h)=f(x,t)
k_2(x,t,f,h)=f(x+h*0.5*k_1(x,t,f,h),t+0.5*h)
k_3(x,t,f,h)=f(x+h*0.5*k_2(x,t,f,h),t+0.5*h)
k_4(x,t,f,h)=f(x+h*k_3(x,t,f,h),t+0.5*h)

function integrar(f,x0,t0,tf,h)
    
    rk4=Vector{Float64}[[x0]]
    t=Float64[t0]
    
            for i in t0:h:tf
        
                rk4_nueva=x0+h*(1/3*k_1(x0,t0,f,h)+1/6*k_2(x0,t0,f,h)+1/6*k_3(x0,t0,f,h)+1/3*k_4(x0,t0,f,h))
        
                t0=t0+h
                x0=rk4_nueva
        
        
                push!(t,t0)
                push!(rk4,[x0])
            
            end
        
        
         t,hcat(rk4...)'
            
        
    
end


function SIS(x,t)

    S=x[1]
    I=x[2]
    R=x[3]
    N=x[4]
    β=x[5]
    γ=x[6]
    μ=x[7]

    [-β*S*I+μ*(N-S)+ γ*I, β*S*I-γ*I-μ*I,0,0,0,0,0]
end


function SIR(x,t)
    S=x[1]
    I=x[2]
    R=x[3]
    N=x[4]
    β=x[5]
    γ=x[6]
    μ=x[7]
    
    [-β*S*I+μ*(N-S), β*S*I-γ*I-μ*I, γ*I-μ*R,0,0,0,0]
end
```

### Resultados

```julia
x1=[0.8,0.05,0.15,1,0.16,0.071,0.40]

tps2,resultados_sis=integrar(SIS,x1,0,10,0.1);

suceptibles_sis=resultados_sis[:;1]
infectados_sis=resultados_sis[:;2]

plot(tps2,suceptibles_sis,label="Susceptibles")
plot(tps2,infectados_sis,label="Infectados")



xlabel("Tiempo")
ylabel("S(t),I(t)")
title("SIS ")
legend()
```

```julia
x1=[0.8,0.05,0.15,1,0.16,0.071,0.90]

tps3,resultados_sir=integrar(SIR,x1,0,6,0.1);

suceptibles_sir=resultados_sir[:;1]
infectados_sir=resultados_sir[:;2]
inmunes_sir=resultados_sir[:;3]

plot(tps3,suceptibles_sir,label="Susceptibles")
plot(tps3,infectados_sir,label="Infectados")
plot(tps3,inmunes_sir,label="Recuperados(Inmunes)")



xlabel("Tiempo")
ylabel("S(t),I(t),R(t)")
title("SIR")
legend()
```

```julia
plot(tps2,suceptibles_sis,label="Susceptibles - SIS")

plot(tps3,suceptibles_sir,label="Susceptibles - SIR")

xlabel("Tiempo")
ylabel("S(t)")
title("SIS vs SIR // Susceptibles")
legend()
```

### Datos recabados para obtener las condiciones iniciales.

#### Velocidad de Propagación

Existe una velocidad de propagación entre los grupos de perros. 

Se sabe que $\frac{2}{3}$ de los perros suceptibles a infectarse interaccionan con perros ya infectados. Es decir, un 66.66%. $\rightarrow$ $r=0.66$

Otro dato obtenido de la investigación previa es que un 90% de los perros que tienen rabia mueren. $\rightarrow$ $a=0.90$

Además, la población de perros no infectados, suceptibles a infectarse, es del 99.83% $\rightarrow$ $S_0=0.9983$

Con estos datos anteriores tenemos que la constante adimensional es $m=\frac{0.66}{(0.90)(0.9983)}$ $\rightarrow$ $m=0.73$

Con estos datos, podemos obtener la velocidad de propagación que es $c\geq2\sqrt{1-0.73}$ $\rightarrow$ $c\geq1.032$

#### Ondas en el sistema de Reacción-Difusión de Rabia (Sin dinámica de Población)

El hecho de no tomar en cuenta la dinámica de Población en el grupo de perros suceptibles (S) quiere decir que el cambió en su población solo depende de la interacción con los infectados, mientras que la poblacion de perros infectados depende de la interaccion con los suceptibles, que tan rapido se mueran y del término de difusión, es decir:


$$\frac{dS}{dt}=-rIS$$

$$\frac{dI}{dt}=rIS -aI-D\frac{d^2 I}{dx^2}$$


Adimensionalizando esta ecuación al introducir $m=\frac{a}{rS_0}$ y considerando que se propagan como ondas:

$$I=I(x-ct)=I(z)$$

$$S=S(x-ct)=S(z)$$

Y con la regla de la cadena, llegamos a las siguientes ecuaciones en función de la variable z:

$$ cS'= IS$$

$$ I'' - cI + (S-m)I= 0$$

Introducimos las siguientes condiciones de frontera al suponer que muy lejos la población de zorros infectados desaparece, y solo queda la poblacion de zorros suceptibles.


$$S(+ \infty)=1$$


$$I(- \infty)=0$$

Al también suponer que después del paso de la onda no hay fuentes de infeccion ,ya que la mayoría de los casos de rabia son fatales, y que la población de zorros se termina por estabilizar tambien tenmos las condiciones:

$$I(-\infty)=0$$

y 


$$S'(-\infty)=0$$


Estas consideraciones son analizadas en [1] y dan como resultado que c debe complir con que $c\geq2\sqrt{1-m}$


introduciendo la variable $I'=v$, tenemos el sistema:

$$I'=v$$

$$S'=\frac{IS}{c}$$

$$v'=-cv-(S-m)I$$

Ecuaciones que traducimos en una función vectorial para que sean resueltas por el métododo de Runge-Kutta.



#### Defininiendo las ecuaciones 

```julia
function rabies_nb(x,t)
    I=x[1]
    S=x[2]
    v=x[3]


    if I>0 #Si el grupo de infectados es mayor a cero queremos solucionar las ecuaciones propuestas
    
    [v, (I*S)/1.03, -1.03*v-(S-0.73)*I]

else
    
    [0, 0, 0] #Cuando I=0 supongo que la rabia se erradica y por lo tanto I=I'=I''=0.

end

end
    
```

```julia
function rabies_dip(x,t) 
    
    I=x[1]
    S=x[2]
    v=x[3]
  
if I>0
        [v, (0.66*I*S-0.1*S*(1-S/4))/1.03, (-v-(0.66*S-0.90)*I)/0.56]#aqui se propuso una población tope de S0=4
   
    else
    
        [0, -0.1S*(1-S/4)/1.03, 0] #dejamos la segunda entrada diferente de cero ya que poniendo I y sus derivadas iguales a cero
        #este término no se elimina

end


end
```
